# Welcome to CLUE! 
Mr. Boddy is dead! One of the killers is in this room. Your job is to figure out who the killer is. There will be three players for this game. Each of you will take turns rolling the dice. You can type in what room you'd like to go to, in which you can make a suggestion or an accusation. If you roll a 3 or above, you will be able to reach your desired room. The Confidential Case file will be filled at the beginning of the game, and you will be dealt six cards that are not in the Confidental Case. As you gain clues, you will gain more cards, which will help you rule out suspects, murder weapons, and locations until you can successfully make a correct accusation! Best of luck!
